```python
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('Datasets\student_scores - student_scores.csv')
```

```python
df.head(10)
```

```python
plt.plot(df)
plt.show()
```

```python
plt.scatter(df['Hours'], df['Scores'])
plt.title('Hours v/s Scores')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.grid(True)
```

```python
X = df.iloc[:, :-1].values  
y = df.iloc[:, 1].values  
```

```python
X
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
plt.scatter(x_train, y_train)
plt.title('(Trained) Hours v/s Scores')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.grid(True)
```

```python
reg = LinearRegression()
```

```python
reg.fit(x_train, y_train)
```

```python
line = reg.coef_*X+reg.intercept_

# Plotting for the test data
plt.scatter(X, y)
plt.plot(X, line);
plt.show()
```

```python
a = np.array([[9.25]])
b = reg.predict(a)
```

```python
b
```

```python
plt.scatter(x_train, y_train)
plt.scatter(a, b)
plt.title('(Trained) Hours v/s Scores')
plt.legend(['Actual', 'Predicted'])
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.grid(True)
```

```python
y_pred = reg.predict(x_test)
```

```python
dff = pd.DataFrame({"Actual": y_test,"Predicted": y_pred})
```
