```python
df = pd.read_csv('Datasets\covid_19_india.csv')
```

```python
df
```

```python
df.groupby('State')['Cured','Deaths','Confirmed'].agg('sum')
```

```python
df.groupby('State')['Cured','Deaths','Confirmed'].agg(['sum']).plot(kind = 'bar')
plt.show()
```

```python
df.groupby('State')['Cured'].agg(['sum']).plot(kind = 'bar')
plt.legend(['Cured'])
plt.grid('True')
plt.show()
```

```python
df.groupby('State')['Deaths'].agg(['sum']).plot(kind = 'bar')
plt.legend(['Deaths'])
plt.grid('True')
plt.show()
```

```python
df.groupby('State')['Confirmed'].agg(['sum']).plot(kind = 'bar')
plt.legend(['Confirmed'])
plt.grid('True')
plt.show()
```

```python
df.groupby('Date')['Confirmed'].agg(['sum']).plot(kind = 'line')
plt.title('Corona Wave')
plt.legend(['Confirmed'])
plt.grid('True')
plt.show()
```

```python
df.groupby('Date')['Cured'].agg(['sum']).plot(kind = 'line')
plt.title('Corona Wave')
plt.legend(['Cured'])
plt.grid('True')
plt.show()
```

```python
df.groupby('Date')['Deaths'].agg(['sum']).plot(kind = 'line')
plt.title('Corona Wave')
plt.legend(['Deaths'])
plt.grid('True')
plt.show()
```
